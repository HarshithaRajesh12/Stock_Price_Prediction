# Sentiment Analysis Framework for Stock Price Prediction
## Based on Literature and Proposed Integration (IJSRET Paper)

### Introduction

Sentiment analysis plays a significant role in stock market forecasting by capturing the emotional and psychological aspects of market participants. Prior research indicates that integrating sentiment from news articles and social media can improve predictive models, especially when combined with deep learning architectures like LSTM.

### Literature Review

Several studies in recent years have explored sentiment-enhanced stock prediction frameworks:

- Prior research shows that models combining numerical time-series data with sentiment scores extracted from textual data (BERT, VADER, TextBlob) can outperform models using only price history.

- Surveys indicate that deep learning models that incorporate sentiment indicators (e.g., investor mood from news) often exhibit improved forecasting accuracy compared to standalone price-only models.

### Proposed Hybrid Architecture

A sentiment-enhanced stock prediction model would consist of the following:

1. **Data Collection:** Gather historical closing prices and textual news data.
2. **Sentiment Scoring:** Apply NLP techniques to compute sentiment scores for each date.
3. **Feature Merging:** Combine daily sentiment scores with price features.
4. **Model Training:** Use an LSTM model with multivariate inputs (price + sentiment).
5. **Evaluation:** Compare performance with price-only models.

This approach stems from evidence in the literature that sentiment capture can add predictive power by modelling market psychology.

### Conclusion

This project implements stock price forecasting using RNN and LSTM models trained on historical market data. The results show that deep learning models, especially LSTM, effectively capture time-dependent patterns in financial data.

Although sentiment analysis is not implemented in this notebook, it is discussed in the literature review as a key enhancement for improving prediction accuracy. Future work will focus on integrating sentiment-based features with the LSTM model to build a more robust hybrid forecasting system.
